### Variables & Identifiers

VARIABLES : Container for storing data , no need to declare the type (dynamic)

IDENTIFIER : Name of the variable, function, class
Naming rules :
	•   meaningful
	•   a-z, A-Z, 0-9 & underscore
	•   can't start with a number
	•   space is not allowed
	•   no length restriction
	                  
	• Camel case - myName
	• Pascal case - MyName
	• Snake case - my_name

	• Reserved words can't be used (keywords)
	• Case sensitive
	• keyword.kwlist


  A         →  normal var
  
  _B         → protected
  
  __C       → private

### Memory Block

```python
a = 22
b = 34
c = 22

print(id(a))# memory address
print(id(b))
print(id(c)) # same as a
```

Memory is assigned to the value of variable passed not the variable name


```python
a = "Sunny"
b = 34
c = "Sunny"

print(id(a), id(b), id(c))
```
