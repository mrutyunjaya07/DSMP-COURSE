### Operator

a symbol that performs certain operations

1.     ARITHMETIC OPERATOR
           ADDITION    +
           SUBSTRACTION     -
           MULTIPLICATION   *
           DIVISION   /         (quotient)
           MODULOUS %     (remainder)
           FLOOR DIVISION   //      (quotient)
           EXPONENTIAL   **


2.     RELATIONAL OPERATOR
    	LESS THAN <
    	GREATER THAN >
    	LESSTHAN EQUAL TO  <=
        GREATERTHAN EQUAL TO  >=

3.     EQUALITY OPERATOR   (never rises error)
   
    	EQUAL TO  ==
        NOT EQUAL TO  !=

Operator | uses | Meaning
- | - | -
is | A == B => | return True if A and B are same   
is not | A != B | returns True if A and B are not same

4.   LOGICAL OPERATOR
	 - and  (if both are true then true otherwise false)
	 - or (if either one is true then true otherwise false)
	 - not (not True is False)
	
	for non-boolean values
    
    	- 0 → False
    	- non-zero values →  True  
    	- empty string is always →  False
	
	x and y :
	   x → False return x otherwise y
	
	x or y :
	   x → True return x otherwise y 
	


	 
5.   BITWISE OPERATOR  (deals with the bits 0,1)
	 - &  (b-and) : if both bits are 1 then 1 otherwise 0
	 - |  (b-or) : if at least one  of the bits is 1 then 1 otherwise o
	 - ^  (b-xor) : if both bits are different then 1 else 0
	 - ~  (b-compliment) 
	 - << (b-left shift)    :  right places filled with zero  or  X * 2^n
     -     >> (b-right shift) :  left places filled with zero  or  X / 2^n
	   




6.     ASSIGNMENT OPERATOR
          We can assign values to variable ( = )

    - x += 10  means  x = x + 10
      similarly
    - -= , *= , /= , //= , %=



```python
# multiple assignment
a, b, c, yes = 10, 20.2, "Sunny", True
```

7.     TERNARY OPERATOR
	
    - X = FIRST VALUE if CONDITION else SECOND VALUE  
                   (if condition is true FIRST VALUE or SECOND VALUE)

    - X = 1ST VAL  if  cond1 else 2ND VALUE if cond2 else cond3

8. IDENTITY OPERATOR
    Used for address comparison 
    - is   (if both address are same)
    - is not (if not)




      
9. MEMBERSHIP OPERATOR
    - in → returns true if the object is present in the specified collection 
    - not in → returns true if the object is not present in the specified collection
