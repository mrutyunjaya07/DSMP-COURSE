### String

String is immutable
 - sequence of characters
 - String Literals : string with quotation mark (either ' or ")
 - The quotes surrounding a string are called Delimiters
 - '      '
 - "      "
 - """  multi line string  """

#### indexing

- for n characters in a string, index range is : 0 to n-1  ( left to right )
- each character can be accessed by name[index]
- negative index range is : -n to -1 ( left to right )




```python
# String indexing
#       01234
#       -5-4-3-2-1
name = "Sunny"


print(name[0])
print(name[-5])
```

#### String Slicing

- extracting a part of a string (substring)
- string[ start : end : step ]
- The starting index of the substring (inclusive)
- The ending index of the substring (exclusive)
- The step size or the interval between each character in the substring.
- If omitted, the default is 1
- non-existent indices are ignored

--> [ : 5 ]  ---- [0:5]

--> [ 5 : ]  -----  [ 5 : n ]

- positive index = len(string) + negative index
- Strings Are Immutable
- "    ".lower( )
- string.upper(  )
-  .title()
-  " ".join(string)
-  .count("")

- str1 + str2     → concatenate
- str1 * 2      →   str1 + str1


```python
# String slicing 
#       01234
#       -5-4-3-2-1
name = "Sunny"

name[1:] # [starting: ending]
```

```python
name[1:3] # This means that ending is excluded
```

```python
name[-4:-1]
```

```python
# Palindrome sequence 
# madam # read from left or right it will be same
# malayalam
# nayan
# bob
# Radar

data = input("input a string: ")
reverse = data[::-1] # start: end: step

print(data, reverse)
```

#### Split operation

```python
name = "Sunny Bhaveen Chandra"
name.split(" ")
name.split("a")
```

#### remove extra spaces front and back of the string

```python
name = "  Sunny   "
print(name.strip(" "))

name1 = "  Sunny   "
print(name1.lstrip(" "))

name2 = "Sunny   "
print(name2.rstrip(" "))

print(name, name1, name2) # doesn't change the original string # creates a copy
```
